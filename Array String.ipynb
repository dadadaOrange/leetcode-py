## 1047. Remove All Adjacent Duplicates In String
[1047. Remove All Adjacent Duplicates In String](https://leetcode.com/problems/remove-all-adjacent-duplicates-in-string/)

```python
class Solution:
    def removeDuplicates(self, s):
        chars = list(s)
        i = 1
        j = 1
        n = len(chars)
        while j < n:
            if i == 0 or chars[i - 1] != chars[j]:
                chars[i] = chars[j]
                i += 1
            else:
                i -= 1
            j += 1
        return "".join(chars[:i])
    
    def removeDuplicatesStack(self, s: str) -> str:
        chars = list(s)
        stack = deque(chars[0])
        i = 1
        while i < len(chars):
            if len(stack) > 0 and stack[-1] == chars[i]:
                stack.pop()
            else:
                stack.append(chars[i])
            i += 1

        # post processing
        result = "".join(stack)
        return result
```

## 1209. Remove All Adjacent Duplicates in String II
[1209. Remove All Adjacent Duplicates in String II](https://leetcode.com/problems/remove-all-adjacent-duplicates-in-string-ii/)

```python
from collections import deque
class Solution:
    def removeDuplicates(self, s, k):
        # corner case
        if len(s) <= k:
            return s
        # linear scan the chars, use the stack to compare the top element in the stack with curr pointer of chars
        chars = list(s)
        count = deque()   
        i = 0
        while i < len(chars):
            if i == 0 or chars[i] != chars[i - 1]:
                count.append(1)
            else:
                cur_count = count.pop() + 1
                if cur_count == k:
                    # remove last k element in chars
                    del chars[i - k + 1 : i + 1]
                    i = i - k
                else:
                    count.append(cur_count)
            i += 1
        return "".join(chars)
```

## 1099. Two Sum Less Than K
[1099. Two Sum Less Than K](https://leetcode.com/problems/two-sum-less-than-k/)

```python
def twoSumLessThanK(self, nums, k):
    nums.sort()
    res = -1
    i = 0
    j = len(nums) - 1
    while i < j:
        cur_sum = nums[i] + nums[j]
        if cur_sum >= k:
            j -= 1
        else:
            if cur_sum > res:
                res = cur_sum
            i += 1
    return res
```

## 1679. Max Number of K-Sum Pairs
[1679. Max Number of K-Sum Pairs](https://leetcode.com/problems/max-number-of-k-sum-pairs/)

```python
def maxOperations(self, nums, k):
    # sort the array
    nums.sort()
    i = 0
    j = len(nums) - 1
    count = 0
    while i < j:
        cur_sum = nums[i] + nums[j]
        if cur_sum == k:
            count += 1
            i += 1
            j -= 1
        elif cur_sum < k:
            i += 1
        else:
            j -= 1
    return count

# time: O(nlogn) + O(n)
# space: O(logn)         --> bottle neck is sort operation
```

## 15. 3Sum
[15. 3Sum](https://leetcode.com/problems/3sum/)

```python
# list[list[]]
res = []
res.append([1, 2])
print(res)
res.append([3, 4])
print(res)
```

```python
# deduplication by using while loop to jump the same elements in the list
# sort first
class Solution:
    def threeSum(self, nums):
        # self.res = set()
        self.res = []
        nums.sort()
        n = len(nums)
        for i in range(n - 2):
            if nums[i] > 0:
                break
            if i == 0 or nums[i] != nums[i - 1]:
                self.twoSum(nums, nums[i], i + 1, n - 1)
        return self.res
    
    def twoSum(self, nums, i_sum, l, r):
        while l < r:
            cur_sum = nums[l] + nums[r]
            if cur_sum == -1 * i_sum:
                # self.res.add((i_sum, nums[l], nums[r]))
                self.res.append([i_sum, nums[l], nums[r]])
                l += 1
                while l < r and nums[l] == nums[l - 1]:
                    l += 1
                r -= 1
                while l < r and nums[r] == nums[r + 1]:
                    r -= 1
            elif cur_sum < -1 * i_sum:
                l += 1
            else:
                r -= 1   
    # time: O(nlogn + n^2)
    # space: O(logn)
```

## 16. 3Sum Closest
[16. 3Sum Closest](https://leetcode.com/problems/3sum-closest/)

```python
class Solution:
    def threeSumClosest(self, nums, target):
        nums.sort()
        self.diff = float('inf')
        n = len(nums)
        for i in range(n - 2):
            left = i + 1
            right = n - 1
            self.twoSum(nums, i, left, right, target)
        return target - self.diff
    
    def twoSum(self, nums, i, l, r, target):
        while l < r:
            cur_sum = nums[l] + nums[r] + nums[i]
            if abs(target - cur_sum) < abs(self.diff):
                self.diff = target - cur_sum
            if cur_sum > target:
                r -= 1
            else:
                l += 1
            if self.diff == 0:
                return
        return
```

## 259. 3Sum Smaller
[259. 3Sum Smaller](https://leetcode.com/problems/3sum-smaller/)

```python
class Solution:
    """ 
      0 1 2 3 4 5 6
    [-2,0,1,1,1,1,3] target = 2
      i
        l
                  r
    """
    def threeSumSmaller(self, nums, target):
        n = len(nums)   
        if n <= 2:
            return 0
        nums.sort()
        res = 0
        for i in range(n - 2):
            res += self.twoSum(nums, target - nums[i], i + 1, n - 1)
        return res
    
    def twoSum(self, nums, target, l, r):
        count = 0
        while l < r:
            cur_sum = nums[l] + nums[r]
            if cur_sum >= target:
                r -= 1
            else:
                # if current r is smaller than target, all the elements at r's left are < target
                count += r - l
                l += 1
        return count
```

```python
class Solution:
    def fourSum(self, nums, target):
        # corner case
        self.res = set()
        n = len(nums)
        if n <= 3:
            return self.res
        nums.sort()
        for i in range(n - 3):
            if i == 0 or nums[i] != nums[i - 1]:
                for j in range(i + 1, n - 2):
                    if j == i + 1 or nums[j] != nums[j - 1]:
                        self.twoSum(nums, nums[i], nums[j], j + 1, n - 1, target)
        return self.res
    
    def twoSum(self, nums, a, b, l, r, target):
        target = target - a - b
        while l < r:
            cur_sum = nums[l] + nums[r]
            if cur_sum == target:
                self.res.add((a, b, nums[l], nums[r]))
                l += 1
                while l < r and nums[l] == nums[l - 1]:
                    l += 1
                r -= 1
            elif cur_sum < target:
                l += 1
            else:
                r -= 1

solu = Solution()
print(solu.fourSum([1, 2, 3, 4], 10))
```

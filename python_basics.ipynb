## Linkedlist

```python
class ListNode():
    # constructor
    def __init__(self, value):
        self.value = value
        self.next = None
        
# initiate class of list_node
cici = ListNode(226)
lingzi = ListNode(925)
cici.next = lingzi
print(cici.value)
print(cici.next.value)
```

```python
class LinkedList(object):
    def __init__(self, head=None):
        self.head = head
        
    def append(self, new_element):
        current = self.head
        if self.head:
            while current.next:
                current = current.next
            current.next = new_element
        else:
            self.head = new_element

new_head = LinkedList(cici)
print(new_head.head.value)
```

## Stacks

```python
from collections import deque
a = deque([1, 2, 3])
print(a)

# stacks

# push
a.append(5) 
print('new stack', a)

# pop
pp = a.pop() 
print('a.pop()', a, 'pop item', pp)

# peek
top_ele = a[-1]
print('peek the top element', top_ele)

# is_empty
len_stack = len(a)
print('is empty', len_stack == 0)
```

## Queue

```python
from collections import deque

# queue 
a = deque([1, 2, 3, 'cici'])
print('original queue', a)

# enqueue
a.append(5)
print('new queue', a)

# dequeue
ele_deque = a.popleft()
print('new queue', a, 'dequeue element', ele_deque)

# peek
peek = a[0]
print('peek', peek)

# is empty
len_q = len(a)
print('is empty', len_q == 0)
```

## Loop

- break continue just used for the current loop
- break:  break the current loop and jump to the outer loop
- continue: skip the current condition and stay in the same loop

```python
a = [1, 2, 3, 4]
for item in a:
    print(item)
    
i = 1
j = 0
while (i < len(a) and i >= 0):
    while j < i:
        print("test break")
        break
    print(a[i])
    i += 1
```

## PriorityQueue

```python
import heapq
li = [5, 7, 9, 1, 3]
heapq.heapify(li)

# create heap
print("heapify: ",li)

# push
heapq.heappush(li, 4)
print("heappush: ", li)

# pop
ele = heapq.heappop(li)
print("pop element is :", ele)
print("heappop: ", li)

# k largest
l2 = [6, 7, 9, 4, 3, 5, 8, 10, 1]
heapq.heapify(l2)
print(l2)
print("3 largest are :", heapq.nlargest(3, l2))
# k smallest
print("3 smallest are :", heapq.nsmallest(3, l2))
```

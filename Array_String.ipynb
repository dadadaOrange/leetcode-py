## 125. Valid Palindrome
[125. Valid Palindrome](https://leetcode.com/problems/valid-palindrome/)

```python

# two pointers: only check alpha-num characters
def isPalindrome(s):
    n = len(s)
    if n <= 1:
        return True
    l = 0
    r = n - 1
    while l < r:
        # find the left alnum
        while l < r and not s[l].isalnum():
            l += 1
        # find the right alnum
        while l < r and not s[r].isalnum():
            r -= 1
        # compare
        if s[l].lower() != s[r].lower():
            return False
        l += 1
        r -= 1
    return True

s = "abc,,,d   cBA"
print(isPalindrome(s))

        
```

## 680. Valid Palindrome II
[680. Valid Palindrome II](https://leetcode.com/problems/valid-palindrome-ii/)

```python
# Two pointers
class Solution:
    """
    flag = true -> false
        abca
        i
           j
    """
    def validPalindrome(self, s: str) -> bool:
        n = len(s)
        if n <= 1:
            return True
        l = 0
        r = n - 1
        while l < r:
            if s[l] != s[r]:
                return self.pali(s, l + 1, r) or self.pali(s, l, r - 1)
            l += 1
            r -= 1
        return True
    
    # check
    def pali(self, s, l, r):
        while l < r:
            if s[l] != s[r]:
                return False
            l += 1
            r -= 1
        return True
```

```python
# Recursion
def check(s, l, r, flag):
    # base case
    if l >= r:
        return True
    # current level
    if s[l] == s[r]:
        return self.check(s, l + 1, r - 1, flag)
    else:
        if flag:
            return check(s, l + 1, r, False) or check(s, l, r - 1, False)
        else:
            return False
```

## 1047. Remove All Adjacent Duplicates In String
[1047. Remove All Adjacent Duplicates In String](https://leetcode.com/problems/remove-all-adjacent-duplicates-in-string/)

```python
class Solution:
    def removeDuplicates(self, s):
        chars = list(s)
        i = 1
        j = 1
        n = len(chars)
        while j < n:
            if i == 0 or chars[i - 1] != chars[j]:
                chars[i] = chars[j]
                i += 1
            else:
                i -= 1
            j += 1
        return "".join(chars[:i])
    
    def removeDuplicatesStack(self, s: str) -> str:
        chars = list(s)
        stack = deque(chars[0])
        i = 1
        while i < len(chars):
            if len(stack) > 0 and stack[-1] == chars[i]:
                stack.pop()
            else:
                stack.append(chars[i])
            i += 1

        # post processing
        result = "".join(stack)
        return result
```

## 1209. Remove All Adjacent Duplicates in String II
[1209. Remove All Adjacent Duplicates in String II](https://leetcode.com/problems/remove-all-adjacent-duplicates-in-string-ii/)

```python
class Solution:
    """
        two pointers + stack
        [0, i) keep
        j: is processing
        stack contians the number of elememts        
    """
    def removeDuplicates(self, s, k):
        arr = list(s)
        stack = deque()
        # initial
        stack.append(1)
        i = 1
        j = 1
        while j < len(arr):
            if i == 0 or arr[i - 1] != arr[j]:
                arr[i] = arr[j]
                stack.append(1)
            else:
                count = stack.pop() + 1
                if count == k:
                    # remove
                    i = i - k
                else:
                    arr[i] = arr[j]
                    stack.append(count)
            i += 1
            j += 1
        return "".join(arr[: i])
```

```python
from collections import deque
class Solution:
    def removeDuplicates(self, s, k):
        # corner case
        if len(s) <= k:
            return s
        # linear scan the chars, use the stack to compare the top element in the stack with curr pointer of chars
        chars = list(s)
        count = deque()   
        i = 0
        while i < len(chars):
            if i == 0 or chars[i] != chars[i - 1]:
                count.append(1)
            else:
                cur_count = count.pop() + 1
                if cur_count == k:
                    # remove last k element in chars
                    del chars[i - k + 1 : i + 1]
                    i = i - k
                else:
                    count.append(cur_count)
            i += 1
        return "".join(chars)
```

```python
def twoSumLessThanK(self, nums, k):
    nums.sort()
    res = -1
    i = 0
    j = len(nums) - 1
    while i < j:
        cur_sum = nums[i] + nums[j]
        if cur_sum >= k:
            j -= 1
        else:
            if cur_sum > res:
                res = cur_sum
            i += 1
    return res
```

## 16. 3Sum Closest
[16. 3Sum Closest](https://leetcode.com/problems/3sum-closest/)

```python
class Solution:
    def threeSumClosest(self, nums, target):
        nums.sort()
        self.diff = float('inf')
        n = len(nums)
        for i in range(n - 2):
            left = i + 1
            right = n - 1
            self.twoSum(nums, i, left, right, target)
        return target - self.diff
    
    def twoSum(self, nums, i, l, r, target):
        while l < r:
            cur_sum = nums[l] + nums[r] + nums[i]
            if abs(target - cur_sum) < abs(self.diff):
                self.diff = target - cur_sum
            if cur_sum > target:
                r -= 1
            else:
                l += 1
            if self.diff == 0:
                return
        return
```

## 259. 3Sum Smaller
[259. 3Sum Smaller](https://leetcode.com/problems/3sum-smaller/)

```python
class Solution:
    """ 
      0 1 2 3 4 5 6
    [-2,0,1,1,1,1,3] target = 2
      i
        l
                  r
    """
    def threeSumSmaller(self, nums, target):
        n = len(nums)   
        if n <= 2:
            return 0
        nums.sort()
        res = 0
        for i in range(n - 2):
            res += self.twoSum(nums, target - nums[i], i + 1, n - 1)
        return res
    
    def twoSum(self, nums, target, l, r):
        count = 0
        while l < r:
            cur_sum = nums[l] + nums[r]
            if cur_sum >= target:
                r -= 1
            else:
                # if current r is smaller than target, all the elements at r's left are < target
                count += r - l
                l += 1
        return count
```

```python
# Quick Sort
class Solution:
    import random
    def partition(self, nums, left, right, pivot_idx):
        pivot = nums[pivot_idx]
        cur_pos = right
        # swap the pivot to right
        nums[pivot_idx], nums[right] = nums[right], nums[pivot_idx]
        right = right - 1
        while left <= right:
            if nums[left] < pivot:
                left += 1
            else:
                nums[left], nums[right] = nums[right], nums[left]
                right -= 1
        # put the pivot to the right position
        nums[left], nums[cur_pos] = nums[cur_pos], nums[left]
        return left
    def quickselect(self, nums, left, right):
        # base case
        if left < right:
            # find pivot
            pivot_idx = random.randint(left, right)
            position = self.partition(nums, left, right, pivot_idx)
            
            self.quickselect(nums, left, position - 1)
            self.quickselect(nums, position + 1, right)

    def sortArray(self, nums):
        self.quickselect(nums, 0, len(nums) - 1)
        return nums
```

## 88. Merge Sorted Array
[88. Merge Sorted Array](https://leetcode.com/problems/merge-sorted-array/)

```python
class Solution:
    def merge(self, nums1, m, nums2, n):
        """
        Do not return anything, modify nums1 in-place instead.
        """
        i = m - 1
        j = n - 1
        k = m + n - 1
        while i >= 0 and j >= 0:
            if nums1[i] >= nums2[j]:
                nums1[k] = nums1[i]
                i -= 1
            else:
                nums1[k] = nums2[j]
                j -= 1
            k -= 1
        if j >= 0:
            nums1[0 : k + 1] = nums2[0: j + 1]
```

```python
class Solution:
    def strStr(self, haystack: str, needle: str):
        m = len(haystack)
        n = len(needle)
        for i in range(m - n + 1):
            match = True
            for j in range(n):
                if haystack[i + j] != needle[j]:
                    match = False
            if match:
                return i
        return -1
```

## 443. String Compression
[443. String Compression](https://leetcode.com/problems/string-compression/)

```python
class Solution:
    def compress(self, chars):
        """
         a a a a b b b b d d d d 
           i
                 j'
         j
         
         [0, i)  keep
         [j, j'] same letters
         if j' - j = 1
         else:
            while j' - j -> tostring -> append one by one
        """
        i = 0
        j = 0
        k = 0
        n = len(chars)
        while k < n:
            while k < n and chars[k] == chars[j]:
                k += 1
            count = k - j
            chars[i] = chars[j]
            i += 1
            # check the count and append to i position
            if count > 1:
                for ch in str(count):
                    chars[i] = ch
                    i += 1
            # move j
            j = k
        return i
```

```python
# expand right [l, r] -> uniqle substring
class Solution:
    def lengthOfLongestSubstring(self, s):
        n = len(s)
        if n <= 1:
            return n
        dup = set()
        l = 0
        global_max = 0
        for r in range(n):
            while s[r] in dup:
                # shrink left to make [i, j] valid
                dup.remove(s[l])
                l += 1
            # satisfy
            global_max = max(global_max, r - l + 1)
            # add current
            dup.add(s[r])
        return global_max
```

```python
class Solution:
    def findAnagrams(self, s, p):
        # counter to save p
        res = []
        p_dict = Counter(p)
        # default dict to save the sliding window elements in s
        m = len(s)
        n = len(p)
        i = 0
        j = n - 1
        s_dict = Counter(s[i: j + 1])
        while j < m:
            if p_dict == s_dict:
                res.append(i)
            # move the window to right
            if s_dict[s[i]] <= 1:
                del s_dict[s[i]]
            else:
                s_dict[s[i]] -= 1
            i += 1
            j += 1
            if j < m:
                s_dict[s[j]] += 1
        return res
```

## 485. Max Consecutive Ones
[485. Max Consecutive Ones](https://leetcode.com/problems/max-consecutive-ones/)

```python
class Solution:
    def findMaxConsecutiveOnes(self, nums):
        global_max = 0
        dp = 0
        for i in range(len(nums)):
            if nums[i] == 1:
                dp += 1
                global_max = max(global_max, dp)
            else:
                dp = 0
        return global_max
```

## 487. Max Consecutive Ones II
[487. Max Consecutive Ones II](https://leetcode.com/problems/max-consecutive-ones-ii/)

```python
class Solution:
    def findMaxConsecutiveOnes(self, nums):
        n = len(nums)
        if n <= 1:
            return n
        i = 0
        j = 0
        global_max = 0
        flag = True
        while j < n:
            if nums[j] == 0:
                if flag:
                    global_max = max(global_max, j - i + 1)
                    flag = False
                else:
                    # move i
                    while i <= j and nums[i] != 0:
                        i += 1
                    i += 1
            else:
                global_max = max(global_max, j - i + 1)
            j += 1
        return global_max
```

```python
# Expand right
class Solution:
    def longestOnes(self, nums: List[int], k: int) -> int:
        n = len(nums)
        if n <= k:
            return n
        l = 0
        global_max = 0
        for r in range(n):
            # expand r
            if nums[r] == 0:
                k -= 1

            if k < 0:
                # move left to satisify k
                while nums[l] != 0:
                    l += 1
                l += 1
                k += 1
            # update global_max
            global_max = max(global_max, r - l + 1)
        return global_max
```
